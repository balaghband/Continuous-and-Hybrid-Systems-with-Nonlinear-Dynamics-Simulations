###Coupled Van Der Pol Benchmark

We use solve ivp from scipy.integrate library for solving nonlinear continuous systems so, we import this library in addition to other necessary libraries.

```python
from scipy.optimize import fsolve
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import matplotlib.animation as animation
import math
from scipy.integrate import odeint
```

$$
\begin{aligned}
    \dot{x}_1 &= y_1 \\
    \dot{y}_1 &= \mu(1 - x_1^2)y_1 + b(x_2 - x_1) - x_1 \\
    \dot{x}_2 &= y_2 \\
    \dot{y}_2 &= \mu(1 - x_2^2)y_2 - b(x_2 - x_1) - x_2 \\
    \dot{b} &= 0
\end{aligned}
$$

 This model is for a reachable set overapproximation s = x + y + z versus time with
 the value of Âµ as 1. We used random initial conditions for each x1 and x2 between 1.25
 to 1.55, for y2 between 2.35 and 2.45, and for b between 1 and 3. All boundaries are
 included in the sets.

```python
mu = 1
initial_conditions_coupled =[]
for i in range(30):
    x_1_0 = np.random.uniform(1.25,1.55)
    x_2_0 = np.random.uniform(1.25,.55)
    y_1_0 = np.random.uniform(2.35,2.45)
    y_2_0 = np.random.uniform(2.35,2.45)
    b = np.random.uniform(1,3)
    initial_conditions_coupled.append([x_1_0,x_2_0,y_1_0,y_2_0,b])




def coupled_van_der_pol_benchmark(t,state):
    x_1,x_2,y_1,y_2,b = state
    x_1_dot = y_1
    y_1_dot = mu*(1 - (x_1**2))*y_1 + b*(x_2 - x_1) - x_1
    x_2_dot = y_2
    y_2_dot  = mu*(1 - (x_2**2))*y_2 - b*(x_2 - x_1) - x_2
    b_dot = 0
    return [x_1_dot,x_2_dot,y_1_dot,y_2_dot,b_dot]

```

This code below shows the change of x1 over time with 30 random initial conditions.
 As we see, the behavior of the x1 variable is the same for each initial condition during
 time and it is the reason for the thickness of the blue line since it consists of 30 blue lines
 (x1 values) with different initial conditions. For plotting this figure, we used the first
 item in the output set of the model and also the model time.

```python
plt.figure(figsize=(12, 8))
for i,initial_condition in enumerate(initial_conditions_coupled):
    coupled_solution  = solve_ivp(coupled_van_der_pol_benchmark,[0,20],initial_condition,method='RK45',t_eval=np.linspace(0,20,1000))
    plt.plot(coupled_solution.t,coupled_solution.y[0], color = 'blue')

plt.xlabel("time")
plt.ylabel("$x_1")
plt.show()

```

This figure shows the change of y1 over time with 30 random initial conditions.
